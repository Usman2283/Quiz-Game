```python
#Quiz Game
questions = [
    ("What is 2 + 2?", "4"),
    ("Capital of Pakistan?", "Islamabad"),
    ("5 + 6?", "11")
]
score = 0
for q, ans in questions:
    user = input(q +  " ")
    if user.lower() == ans.lower():
        print("Correct")
        score += 1
    else:
        print("Wrong")
print("Score:", score, "/", len(questions))
```
